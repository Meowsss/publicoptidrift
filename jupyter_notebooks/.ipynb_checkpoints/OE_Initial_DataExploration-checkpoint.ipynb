### Goals
The first goal for this project is to predict when CH2CDWFLO is inaccurate. The dataset includes periods where the sensor is out of calibration. We will consider CH1CDWFLO to be reasonably accurate and that CH2CDWFLO, when not out of calibration, should be within 10% of CH1CDWFLO. Looking at the long view of the data, I see that CH1CDWFLO might be slipping slowly over the years. Let me know if you see any notable patterns on that front.

```python
def load_all_data(rootdir): 
    if not os.path.exists(rootdir):
        print('The specified rootdir does not exist') 
    assert type(rootdir) == str, 'Input must be a string'
    file_list = [f for f in glob.glob(os.path.join(rootdir, '*.csv'))]
    
    d = {}
    #initiate dictionary to store dataframes in
    count = 0
    for file in file_list:
        count += 1
        name = os.path.split(file)[1].split('Clems ')[1].split('.')[0]
        #takes the second part of the filename (after clems) but leaves off the 'csv' part
        data = pd.read_csv(file)
        new_set = {name: data}
        d.update(new_set)
    return d
```

```python
def acc_of_model(withinpercent, df):
    '''withinpercent should be a decimal - 0.1 means within 10% of the actual'''
    array_predict = predict
    array_actual = df['CH1CDWFLO']
    array = abs(array_predict - array_actual)/array_actual
    num = 0
    for i in range(len(array)):
        if array[i] < withinpercent:
        #this is considered accurate- predicted is less than 20% off from actual
            num = num+1
    percent = num / len(df)
#number of accurate predictions divided by number of predictions
    print('The percentage of the predictions that are within ' + str(withinpercent)+ ' of the actual:',percent)
    return 
```

```python
def lasso_reg(traindata, predictors, alpha, models_to_plot={}):
    lassoreg = Lasso(alpha = alpha, normalize = True, max_iter = 1e5)
    lassoreg.fit(traindata[predictors],traindata['CH1CDWFLO'])
    #y is what we are trying to predict
    y_pred_train = lassoreg.predict(traindata[predictors])
    #y_pred_test = lassoreg.predict(testdata[predictors])
    #Check if a plot is to be made for the entered alpha
    #model_to_plot variable is set up as dictionary alpha:position_of_subplot 
    if alpha in models_to_plot: 
        fig = plt.figure(figsize=(10, 10))
        plt.subplot(models_to_plot[alpha])
        plt.tight_layout()
        #x below is what we are plotting to check accuracy - we would use index
        plt.plot(traindata['index'],y_pred_train, 'g.')
        plt.plot(traindata['index'],traindata['CH1CDWFLO'],'r.')
        plt.title('Plot for alpha: %.3g'%alpha)
    
    #Return the result in pre-defined format
    #mse_train = mean_squared_error(traindata['CH1CDWFLO'], y_pred_train)
   # mse_test = mean_squared_error(testdata['CH1CDWFLO'], y_pred_test)
    #mse = sum(mean_squared_error(traindata['CH1CDWFLO'], y_pred_train))
    rss = sum((y_pred_train-traindata['CH1CDWFLO'])**2)
    ret = [rss]
    ret.extend([lassoreg.intercept_])
    #ret.extend([mse_train])
    #ret.extend([mse_test])
    ret.extend(lassoreg.coef_)
    return ret
```
